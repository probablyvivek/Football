```python
from pywaffle import Waffle # PyWaffle Documentation --> https://buildmedia.readthedocs.org/media/pdf/pywaffle/latest/pywaffle.pdf
import matplotlib.pyplot as plt #Matplotlib pyplot to plot the charts
import pandas as pd
import matplotlib as mpl
import highlight_text
```

```python
#Import the data into a dataframe
df = pd.read_excel("Top5_European_Leaques_Player_Data.xlsx")
```

```python
data2 = df[['Player','Progressive Passes','Passes in 18 yards', '1/3rds', 'Press passes']].copy()
```

```python
data2['Total'] = data2['Progressive Passes'] + data2['Passes in 18 yards'] + data2['1/3rds'] + data2['Press passes']
```

```python
data2 = data2.sort_values(by=['Total'],ascending=False)
```

```python
data2
```

```python
data2 = data2[:9]
```

```python
data2 = data2.T
data2.columns = data2.iloc[0]
data2 = data2.drop('Player')
data2 = data2.drop('Total')
```

```python
#Divide the data by 50 so it doesn't plot too many boxes
data2 = data2/20

#Create a new dataframe so that it is in the proper set up 
my_list = data2.columns.values.tolist()
data3 = pd.DataFrame(data2,columns=my_list)
```

```python
data3
```

```python
#Create the waffle chart. We use pyplot.figure() witht the figure class = Waffle
#Refer to the documentation to get exact explanation on everything.
fig = plt.figure(figsize=(12,10))
fig = plt.figure(
    FigureClass = Waffle,
    plots={
        '331': {                              #refer matplotlib subplot grids, '331' means 3 x 3 grid, first subplot
            'values': data3.iloc[:,0],
            'title': {
                'label': my_list[0],
                'color': 'white',
                'fontsize': 20,
            },
            },
        '332': {                             
            'values': data3.iloc[:,1],
            'title': {
                'label': my_list[1],
                'color': 'white',
                'fontsize': 20,
            },
            },
        '333': {                           #Line 20   
            'values': data3.iloc[:,2],
            'title': {
                'label': my_list[2],
                'color': 'white',
                'fontsize': 20,
            }
            },
        '334':{                              
            'values': data3.iloc[:,3],
            'title': {
                'label': my_list[3],
                'color': 'white',
                'fontsize': 20,
            },
            }
        ,
        '335': {                              
            'values': data3.iloc[:,4],
            'title': {
                'label': my_list[4],
                'color': 'white',
                'fontsize': 20,
            },
            },
        '336': {
            'values': data3.iloc[:,5],                              
            'title': {
                'label': my_list[5],
                'color': 'white',
                'fontsize': 20,
            },
            },
        '337': {                              
            'values': data3.iloc[:,6],
            'title': {
                'label': my_list[6],
                'color': 'white',
                'fontsize': 20,
            },
            },
        '338': {                             
            'values': data3.iloc[:,7],
            'title': {
                'label': my_list[7],
                'color': 'white',
                'fontsize': 20,
            },
            },
        '339': {
            'values': data3.iloc[:,8],
            'title': {
                'label': my_list[8],
                'color': 'white',
                'fontsize': 20,
            },
            },
    },
    rows=5,
    figsize=(12, 10),
    rounding_rule='floor',
    colors=("#83c5be", "#0077b6", "#ffc8dd",'#e29578'),

    facecolor='#313332',
)

#Different colors so we don't have to type them in every time
background = "#313332"
text_color = 'w'
primary = 'red'
secondary = 'lightblue'
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = text_color


#Create the text and highlight the different points

#title Top Passers in European Top 5 Leagues 
fig.text(.01,1.2,"Top Passers in European Top 5 Leagues",fontsize=30,fontweight='bold',fontfamily='Trebushet MS',color=text_color)

HighlightText(x=0.01, y=2,
            s='<Progressive Passes>\n<Passes in 18 yards>\n<Passes in Attacking 3rd>\n<Passes under Pressure>',
            highlight_textprops=[{"color": '#83c5be'},
                                {"color": '#0077b6'},
                                {"color": '#ffc8dd'},
                                {"color": '#e29578'},],
            ax=fig.axes[0],
            fontsize=20,
            fontname='Trebushet MS',
            color=text_color
            )   
fig.patch.set_facecolor(background)
fig.text(.01,-0.04,"Each block represents roughly 20 passes\nData via fbref",fontstyle='italic',fontsize=12,fontfamily='Trebushet MS',color=text_color)




plt.savefig('waffle.png',dpi=300,bbox_inches = 'tight',facecolor=background)


```
