```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
```

```python
df = pd.read_html("https://fbref.com/en/comps/9/Premier-League-Stats")[0]
```

```python
df.sort_values(by="Rk", ascending=True)
```

```python
df = df.sort_values(by="Rk", ascending=False)
```

```python
plt.style.use("fivethirtyeight")
fig = plt.figure(figsize=(20,18))

facecolor = "#e9ecef"

#set the facecolor of the plot
fig.patch.set_facecolor(facecolor)

#dual axis GF and xG on x axis and Squad on y axis
ax1 = plt.subplot2grid((2,2), (0,0), rowspan=2) # subplot2grid( (rows, columns), (starting point), rowspan, colspan)


#reduce alpha of lines and grid
ax1.grid(True, alpha=0.2)

#spines off
for i in ['top','bottom','left','right']:
    ax1.spines[i].set_visible(False)

ax1.set_facecolor(facecolor)

#plot xG
df.plot(x='xG', y='Squad', kind='scatter', s=400, ax=ax1, color='#e76f51', alpha=0.5, edgecolors='#e76f51', linewidth=3)

#xlabel off
ax1.set_xlabel('')
ax1.set_ylabel('')
ax1.set_xlim(20,100)

ax2 = ax1.twinx() #we have to have a dual axis to plot different y axis values on the same graph 

#plot GF
df.plot(x='GF', y='Squad', s=400, kind='scatter', ax=ax2, color='#7209b7', alpha=0.5, edgecolor='#7209b7', lineWidth=3, legend=False)

ax2.set_xlim(20,100)
for i in ['top','bottom','left','right']:
    ax2.spines[i].set_visible(False)
ax2.grid(False)
ax2.set_ylabel('')
ax2.set_yticks([])

#line between xG and GF dots
for i in range(len(df)):
    if df.iloc[i]['xG'] > df.iloc[i]['GF']:
        ax1.plot([df.iloc[i]['GF'], df.iloc[i]['xG']], [df.iloc[i]['Squad'], df.iloc[i]['Squad']], color='#7209b7', alpha=0.5, linestyle='--', linewidth=2, zorder=1)
    else:
        ax1.plot([df.iloc[i]['GF'], df.iloc[i]['xG']], [df.iloc[i]['Squad'], df.iloc[i]['Squad']], color='#e76f51', alpha=0.5, linestyle='--', linewidth=2, zorder=1)



#plot difference between xG and GF 
    if df.iloc[i]['xG'] > df.iloc[i]['GF']:
        ax1.annotate(str(round(df.iloc[i]['xG']-df.iloc[i]['GF'], 1)), xy=(df.iloc[i]['xG'], df.iloc[i]['Squad']), xytext=(df.iloc[i]['xG']+5, df.iloc[i]['Squad']), color='#7209b7', fontsize=11, fontname ='Trebushet MS', ha='right', va='center', zorder=5)
    else:
        ax1.annotate(str(round(df.iloc[i]['xG']-df.iloc[i]['GF'], 1)), xy=(df.iloc[i]['xG'], df.iloc[i]['Squad']), xytext=(df.iloc[i]['xG']-5, df.iloc[i]['Squad']), color='#e76f51', fontsize=11, fontname='Trebushet MS', ha='left', va='center', zorder=5)

#Title Teams Overperformin #e76f51 while #7209b7 are underperforming
colors = ['#e76f51', '#7209b7']

ax1.set_title('Expected Goals', fontsize=22, fontname='Trebushet MS', fontweight= 'bold', color=colors[0], x = 0.28)

ax1.text(0.445, 1.01, '/', fontsize=30, fontname='Trebushet MS', color='black', ha='center', va='center', transform=ax1.transAxes)

ax2.set_title('Goals Scored', fontsize=22, fontname='Trebushet MS', fontweight ='bold', color=colors[1], x = 0.59)

ax2.text(0.43, .99, 'English Premier League | 2021-2022 Season | Figures as of ' + dt.datetime.now().strftime("%d %b %Y"), fontsize=12, fontname='Trebushet MS', color='black', ha='center', va='center', transform=ax2.transAxes)



#lines to show top 4 and bottom 3 teams
ax1.axhline(y=2.3, color='#f94144', linestyle='--', linewidth=2, alpha=0.3, zorder=0)
ax1.axhline(y=15.7, color='#90be6d', linestyle='--', linewidth=2, alpha=0.3, zorder=0)

#add the logo
im = plt.imread('Logo.png')
newax = fig.add_axes([0.01, 0.9, 0.07, 0.09], anchor='NE', zorder=1, frameon=False)
newax.imshow(im)
newax.axis('off')

plt.savefig('Premier League Stats ' + dt.datetime.now().strftime("%d %b %Y") + '.png', bbox_inches='tight', facecolor=facecolor, dpi = 300)

# #another axis to show W and D and L for each team
# ax3 = plt.subplot2grid((2,2), (0,1), rowspan=2)

# #spines off
# for i in ['top','bottom','left','right']:
#     ax3.spines[i].set_visible(False)

# ax3.set_facecolor(facecolor)

# #plot W D and L as stacked bar chart horizontal
# df.plot(x='Squad', y=['W', 'D', 'L'], kind='barh', stacked=True, ax=ax3, color=['#90be6d', 'Grey', 'Red'], alpha=0.5, edgecolor='black', linewidth=1, legend=False)

# #remove ylabel
# ax3.set_ylabel('')

# #remove yticks
# ax3.set_yticks([])

# ax3.set_title('Expected Goals', fontsize=22, fontname='Trebushet MS', fontweight= 'bold', color=colors[0], x = 0.27)
# ax3.text(1.5, .99, 'English Premier League | 2021-2022 Season | Figures as of ' + dt.datetime.now().strftime("%d %b %Y"), fontsize=12, fontname='Trebushet MS', color='black', ha='center', va='center', transform=ax2.transAxes)

```
